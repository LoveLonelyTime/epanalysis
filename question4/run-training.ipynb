# 训练模型

```python
# 载入模型

from training import EPATrainingApp
import torch
```

```python
# 设置参数

argv = ['trainingApp','--epochs=1000','--batch-size=64']

training_app = EPATrainingApp(argv)

training_app.model.load_state_dict(torch.load('model_weights.pth'))

# 开始训练
# 训练里程会保存在 runs 目录下面，使用 TensorBoard 查看 tensorboard --logdir=runs
training_app.main()
```

```python
# 保存模型

torch.save(training_app.model.state_dict(), 'model_weights.pth')
```
